```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# transpose to make 2 rows and N columns
X = np.random.normal(loc=[1, 2], size=(100, 2)).T

plt.plot(X[0, :], X[1, :], '.')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
# scale x-axis by 1/2
A = np.array([[1/2, 0], [0, 1]])
AX = np.dot(A, X)

plt.plot(AX[0, :], AX[1, :], '.')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
# mirror datapoints along the y-axis
B = np.array([[-1, 0], [0, 1]])
BAX = np.dot(B, AX)

plt.plot(BAX[0, :], BAX[1, :], '.')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
# permute the axes (swap)
C = np.array([[0, 1], [1, 0]])
CBAX = np.dot(C, BAX)

plt.plot(CBAX[0, :], CBAX[1, :], '.')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
# rotate by 45 degrees
theta = np.radians(45)
D = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
DCBAX = np.dot(D, CBAX)

plt.plot(DCBAX[0, :], DCBAX[1, :], '.')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
# matrix multiplication is associative
E = np.dot(D, np.dot(C, np.dot(B, A)))
EX = np.dot(E, X) # same as previous DCBAX

plt.plot(EX[0, :], EX[1, :], '.')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
# matrix multiplication is not commutative
F = np.dot(A, np.dot(B, np.dot(C, D)))
FX = np.dot(F, X) # not the same as DCBAX

plt.plot(FX[0, :], FX[1, :], '.')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```
